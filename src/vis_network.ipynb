## NetworkX

Reference:  

* https://networkx.org/documentation/stable/tutorial.html

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

* `nx.Graph()` creates an empty graph, which is a `networkx.classes.graph.Graph` object. 
    - `nx.DiGraph()`: directed graph
* `G.nodes()` and `G.edges()`: check out existing nodes/edges
    * `G.add_node()`, `G.add_nodes_from()` and `G.add_edge()`, `G.add_edges_from()`: add new nodes/edges
    * `G.number_of_nodes()` and `G.number_of_edges()`
    * Each element of `G.nodes()` and `G.edges()` is a dictionary and can be updated to store attributes such as weight and size
    * `G.remove_node()`, `G.remove_nodes_from()`, `G.remove_edge(`, 
    * Use `G.clear()` to clear current graph
* Use `nx.draw()` to visualize graphs

```python
G = nx.Graph()
G.add_node(1)
G.add_nodes_from(range(2, 6))
G.nodes()
```

Adding edges for nodes that do not exist will automatically add those nodes. Therefore, skipping the previous code chunk is perfectly fine. 

```python
G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 5)
G.edges()
```

```python
nx.draw(G)
nx.draw(G, with_labels=True, node_color="#8A3324")
```

### More customization

Options for `nx.draw()`:  

* "node_color": "#A0CBE2",
* "edge_color": colors,
* "width": 4,
* "edge_cmap": plt.cm.Blues,
* "with_labels": False,

```python
G = nx.fast_gnp_random_graph(n=20, p=0.25)
G.edges()
```

```python
# compute positions for the nodes - `pos` is a dict object that maps nodes to positions
# without specifying `seed`, each run will return a different result
pos = nx.spring_layout(G, seed=23)
pos
```

* Graph settings: [nodes](https://networkx.org/documentation/latest/reference/generated/networkx.drawing.nx_pylab.draw_networkx_nodes.html#networkx.drawing.nx_pylab.draw_networkx_nodes), [edges](https://networkx.org/documentation/latest/reference/generated/networkx.drawing.nx_pylab.draw_networkx_edges.html#networkx.drawing.nx_pylab.draw_networkx_edges), [labels](https://networkx.org/documentation/latest/reference/generated/networkx.drawing.nx_pylab.draw_networkx_labels.html#networkx.drawing.nx_pylab.draw_networkx_labels)
* [Matplotlib color maps](https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html)

    cmaps['Perceptually Uniform Sequential'] = [
                'viridis', 'plasma', 'inferno', 'magma', 'cividis']

    cmaps['Sequential'] = [
                'Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',

                'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn']

```python
nx.draw(
    G,
    with_labels=False,
    pos=pos,
    node_color="#A0CBE2",
    edge_color=range(len(G.edges)),
    edge_cmap=plt.cm.Blues,
    width=4
)
```

## nx-altair: NetworkX meets Altair

```python
import nx_altair as nxa

# Draw the graph using Altair
viz = nxa.draw_networkx(G, pos=pos)

# Show it as an interactive plot!
viz.interactive()
```

```python
# Add weights to nodes and edges
for n in G.nodes():
    G.nodes[n].update({
        "label": n, 
        "weight": np.random.randn()
    })

for e in G.edges():
    G.edges[e]['weight'] = np.random.uniform(1, 10)

# Draw the graph using Altair
viz = nxa.draw_networkx(
    G,
    node_label='label',
    node_color='weight',
    pos=pos,
    width='weight',
    edge_color='black',
    cmap='viridis',
)

# Show it as an interactive plot!
viz.interactive()
```
